## 기출문제 1 럭키 스트레이트
현재 점수 N이 주어지면 럭키 스트레이트를 사용할 수 있는 상태인지 아닌지 알려주어라
- 조건 : 자릿수를 기준으로 점수 N을 반으로 나누어 왼쪽 부분의 각 자릿수의 합과 오른쪽 부분의 각 자릿수의 합이 동일해야 함

### 내 풀이

```python
# 입력 조건
# 첫째 줄에 점수 N이 정수로 주어짐
# 단, N의 자릿수는 항상 짝수
n = input()

num_list = [int(n[i]) for i in range(len(n))]
total_sum = sum(num_list)

left_sum = 0
for i in range(0, int(len(n)/2)):
    left_sum += num_list[i]
    
# 출력 조건
# 럭키 스트레이트를 사용할 수 있다면 LUCKY, 사용할 수 없다면 READY 출력    
result = "READY"
if int(total_sum/2) == left_sum: result = "LUCKY"

print(result)
```

### 정답 코드

```python
n = input()
length = len(n) # 정수값의 총 자릿수
summary = 0

# 왼쪽 부분의 자릿수 합 더하기
for i in range(length // 2):
    summary += int(n[i])
    
# 오른쪽 부분의 자릿수 합 빼기
for i in range(length // 2, length):
    summary -= int(n[i])
    
# 왼쪽 부분과 오른쪽 부분의 자릿수 합이 동일한지 검사
if summary == 0:
    print('LUCKY')
else:
    print('READY')
```

## 기출문제 3 문자열 압축
압축할 문자열 S가 매개변수로 주어질 때, 1개 이상 단위로 문자열을 잘라 압축하여 표현한 문자열 중 가장 짧은 것의 길이를 return하는 함수를 만들어라

### 내 풀이

```python
def solution(s):
    answer = 999999999
    
    # 문자열 길이가 1인 경우 더 이상 압축 불가능하므로 1 반환
    if len(s) == 1:
        answer = 1
        return answer
    
    # 1부터 전체 문자열 길이의 절반까지의 단위로 모든 경우를 압축해본다
    for num in range(1, int(len(s)/2) + 1):
        count = 1
        compressed = ''
        
        # 각 단위로 잘랐을때 고려해야할 문자열 길이
        length = (len(s) // num) * num
        end = length - (2 * num)
        
        for i in range(0, end + 1, num):
            # 마지막 iteration인지 확인
            if i == end:
                if s[i:i + num] == s[i + num:i + num + num]:
                    count += 1
                    compressed += str(count)
                    compressed += s[i:i + num]
                else: 
                    if count == 1: count = ''
                    compressed += (str(count) + s[i:i + num])
                    compressed += s[i + num:i + num + num]
                
            else:    
                if s[i:i + num] == s[i + num:i + num + num]:
                    count += 1
                else:
                    if count == 1: count = ''
                    compressed += str(count)
                    compressed += s[i:i + num]
                    count = 1
        
        # 남은 문자열 더해줌            
        compressed += s[length:]
    
        answer = min(answer,len(compressed))
                
    return answer
```

### 정답 코드
완전 탐색을 이용하여 열쇠를 이동이나 회전시켜서 자물쇠에 끼워보는 작업을 전부 시도해본다
- 자물쇠 리스트의 크기를 넓혀서 계산을 하자!
- 자물쇠 리스트에 열쇠 리스트의 값을 더한 뒤, 자물쇠 부분의 더한 결과 값이 모두 1인지 확인하면 됨
- 2차원 리스트를 90도 회전한 결과를 반환하는 함수를 구현하자

```python
# 2차원 리스트 90도 회전
def rotate_a_matrix_by_90_degree(a):
    n = len(a) # 행 길이 계산
    m = len(a[0]) # 열 길이 계산
    result = [[0] * n for _ in range(m)] # 결과 리스트
    for i in range(n):
        for j in range(m):
            result[j][n - i - 1] = a[i][j]
    return result

# 자물쇠의 중간 부분이 모두 1인지 확인
def check(new_lock):
    # 원래 자물쇠의 길이
    lock_length = len(new_lock) // 3
    for i in range(lock_length, lock_length * 2):
        for j in range(lock_length, lock_length * 2):
            if new_lock[i][j] != 1:
                return False
    return True

def solution(key, lock):
    n = len(lock)
    m = len(key)
    # 자물쇠의 크기를 기존의 3배로 변환
    new_lock = [[0] * (n * 3) for _ in range(n * 3)]
    # 새로운 자물쇠의 중앙 부분에 기존의 자물쇠 넣기
    for i in range(n):
        for j in range(n):
            new_lock[i + n][j + n] = lock[i][j]
            
    # 4가지 방향에 대해서 확인
    for rotation in range(4):
        key = rotate_a_matrix_by_90_degree(key) # 열쇠 90도 회전
        # 넓힌 자물쇠에 대해 모든 부분을 끼워 맞춰 본다
        for x in range(n * 2):
            for y in range(n * 2):
                # 자물쇠에 열쇠를 끼워 넣기
                for i in range(m):
                    for j in range(m):
                        new_lock[x + i][y + j] += key[i][j]
                # 새로운 자물쇠에 열쇠가 정확히 들어맞는지 검사
                if check(new_lock) == True:
                    return True
                # 자물쇠에서 열쇠를 다시 빼서 초기화
                for i in range(m):
                    for j in range(m):
                        new_lock[x + i][y + j] -= key[i][j]
    
    return False
```

## 기출문제 4 뱀
뱀 게임 규칙
- 머리가 이동한 칸에 사과가 있다면, 그 칸의 사과는 없어지고 꼬리는 움직이지 않음 (몸길이가 한 칸 늘어나는 셈)
- 머리가 이동한 칸에 사과가 없다면, 꼬리 칸도 같이 움직임 (즉, 몸길이가 변하지 않음)
- 뱀이 벽 또는 자기 자신의 몸과 부딪히면 게임 끝  
게임이 몇 초만에 끝나는지 계산하라

```python
# 첫째 줄에 보드 크기 N
n = int(input())
# 둘째 줄에 사과 개수 k
k = int(input())
# 다음 k개 줄에 사과의 위치 (행, 열)
apples = []
for _ in range(k):
    apples.append(list(map(int, input().split())))
# 다음 줄에 방향 변환 횟수 L이 주어짐
l = int(input())
# 다음 L개 줄에 방향 변환 정보 (X초 후, 방향) 
rotations = []
for _ in range(l):
    rotations.append(list(input().split()))
    
# 초기 시작 판 만들기
map_value = [[0] * n for _ in range(n)]
# 사과 채워넣기
for i in range(k):
    x = apples[i][0]
    y = apples[i][1]
    map_value[x - 1][y - 1] = 1
    
print(map_value)

def rotate_map(a, rotation, head, tail):
    n = len(a)
    map_result = [[0] * n for _ in range(n)] # 결과 리스트
    head_result = [0, 0] # 머리 리스트
    tail_result = [0, 0] # 꼬리 리스트
    
    if rotation == 'L': # 뱀이 왼쪽으로 90도 회전할 경우, 게임 맵을 시계 방향으로 90도 회전시킴
        for i in range(n):
            for j in range(n):
                map_result[j][n - i - 1] = a[i][j]
        head_result[1] = n - head[0] - 1
        head_result[0] = head[1]
        tail_result[1] = n - tail[0] - 1
        tail_result[0] = tail[1]
        
    else: # 뱀이 오른쪽으로 90도 회전할 경우, 게임 맵을 반시계 방향으로 90도 회전시킴
        for i in range(n):
            for j in range(n):
                map_result[n - j - 1][i] = a[i][j]
        head_result[0] = n - head[1] - 1
        head_result[1] = head[0]
        tail_result[0] = n - tail[1] - 1
        tail_result[1] = tail[0]
        
    return map_result, head_result, tail_result

# 게임 시작
# 초 세기
count = 1
# 먼저 뱀은 맨 위 좌측에 위치, 오른쪽으로 움직임
map_value[0][0] = 2
head = [0, 0]
tail = [0, 0]
flag = True
second = int(rotations[0][0])
temp = 0

while count <= second:

    # 벽과 만나거나 자신과 만나는 경우
    if head[1] + 1 == n or map_value[head[0]][head[1] + 1] == 2:
        print('벽 또는 자신과 만남')
        flag = False
        break
    else:
        # 사과가 없는 경우
        if map_value[head[0]][head[1] + 1] == 0: 
            map_value[head[0]][head[1] + 1] = 2 # 머리 이동
            head = [head[0], head[1] + 1]
            map_value[tail[0]][tail[1] + 1] = 0 # 꼬리 이동
            tail = [tail[0], tail[1] + 1]
            print('사과 없음')
        # 사과가 있는 경우 
        else:
            map_value[head[0]][head[1] + 1] = 2 # 머리 이동
            head = [head[0], head[1] + 1]
            print('사과 있음')
            
    if count == second: # 방향 전환할 시간이 되면
        print('방향 전환')
        map_value, head, tail = rotate_map(map_value, rotations[temp][1], head, tail)
        temp += 1
        second = int(rotations[temp][0])

    count += 1
                
print(count)
```

### 정답 코드
- 맵을 회전시킨다는 접근은 잘못된 접근이었다..

```python
n = int(input())
k = int(input())
data = [[0] * (n + 1) for _ in range(n + 1)] # 맵 정보
info = [] # 방향 회전 정보

# 맵 정보(사과 있는 곳은 1로 표시)
for _ in range(k):
    a, b = map(int, input().split())
    data[a][b] = 1

# 방향 회전 정보 입력
l = int(input())
for _ in range(l):
    x, c = input().split()
    info.append((int(x), c))

# 처음에는 오른쪽을 보고 있으므로(동, 남, 서, 북)
dx = [0, 1, 0, -1]
dy = [1, 0, -1, 0]

def turn(direction, c):
    if c == "L":
        direction = (direction - 1) % 4
    else:
        direction = (direction + 1) % 4
    return direction

def simulate():
    x, y = 1, 1 # 뱀의 머리 위치
    data[x][y] = 2 # 뱀이 존재하는 위치는 2로 표시
    direction = 0 # 처음에는 동쪽을 보고 있음
    time = 0 # 시작한 뒤에 지난 '초' 시간
    index = 0 # 다음에 회전할 정보
    q = [(x, y)] # 뱀이 차지하고 있는 위치 정보(꼬리가 앞쪽)

    while True:
        nx = x + dx[direction]
        ny = y + dy[direction]
        
        # 맵 범위 안에 있고, 뱀의 몸통이 없는 위치라면
        if 1 <= nx and nx <= n and 1 <= ny and ny <= n and data[nx][ny] != 2:
            # 사과가 없다면 이동 후에 꼬리 제거
            if data[nx][ny] == 0:
                data[nx][ny] = 2
                q.append((nx, ny))
                px, py = q.pop(0)
                data[px][py] = 0
            # 사과가 있다면 이동 후에 꼬리 그대로 두기
            if data[nx][ny] == 1:
                data[nx][ny] = 2
                q.append((nx, ny))
                
        # 벽이나 뱀의 몸통과 부딪혔다면
        else:
            time += 1
            break
            
        x, y = nx, ny # 다음 위치로 머리를 이동
        time += 1
        
        if index < l and time == info[index][0]: # 회전할 시간인 경우 회전
            direction = turn(direction, info[index][1])
            index += 1
            
    return time

print(simulate())
```

## 기출문제 5 기둥과 보 설치

- 보는 한쪽 끝부분이 기둥 위에 있거나 또는 양쪽 끝부분이 다른 보와 동시에 연결되어 있어야 함

### 내 풀이

```python
def solution(n, build_frame):
    answer = []
    # build_frame의 원소 [x, y, a, b]
    # x, y는 기둥 또는 보를 설치할 교차점의 좌표
    # a는 구조물의 종류 - 기둥은 0, 보는 1
    # b는 작업의 종류 - 삭제 0, 설치 1
    
    wall = [[3] * (n + 1) for _ in range(n + 1)] # 벽면 길이 n인 2차원 행렬 초기화 (3로 채움)
    
    # 작업 수만큼 반복
    for i in range(len(build_frame)):
        # 벽면 배열 인덱스에 맞도록 주어진 위치 값 조정
        x = n - build_frame[i][1]
        y = build_frame[i][0]
        bo_list = [1, 2]
        col_list = [0, 2]
        
        build_type = build_frame[i][2] # 기둥인지 보인지 확인 
        build = True if build_frame[i][3] == 1 else False # 설치할지 삭제할지 확인

        # 기둥을 설치하는 경우
        if build_type == 0 and build:
            if x == n or wall[x][y - 1] in bo_list or wall[x + 1][y] in col_list: 
                wall[x][y] = 0 if wall[x][y] != 1 else 2 # 기둥과 보가 모두 설치되는 칸인 경우 2 할당
        
        # 보를 설치하는 경우
        elif build_type == 1 and build:
            if wall[x + 1][y] in col_list or wall[x + 1][y + 1] in col_list or (wall[x][y + 1] in bo_list and wall[x][y - 1] in bo_list): 
                wall[x][y] = 1 if wall[x][y] != 0 else 2 # 기둥과 보가 모두 설치되는 칸인 경우 2 할당
        
        # 기둥을 제거하는 경우
        elif build_type == 0 and not build: 
            # 제거하려는 기둥 위에 기둥만 있을 경우 
            if wall[x - 1][y] == 0 and wall[x - 1][y - 1] not in bo_list:
                continue
            # 제거하려는 기둥 위(오른쪽)에 보가 있을 경우    
            if wall[x - 1][y] in bo_list:
                if not (wall[x - 1][y - 1] in bo_list and wall[x - 1][y + 1] in bo_list):
                    continue
            # 제거하려는 기둥 위(왼쪽)에 보가 있을 경우 
            if wall[x - 1][y - 1] in bo_list:
                if not (wall[x - 1][y - 2] in bo_list and wall[x - 1][y] in bo_list):
                    continue
            wall[x][y] = 3 if wall[x][y] != 2 else 1
            
        # 보를 제거하는 경우
        else:
            if wall[x][y] in col_list:
                if wall[x + 1][y] not in col_list and wall[x][y - 1] not in bo_list: # 제거하려는 보 위에 기둥만 남게 되는 경우
                    continue
            if wall[x][y + 1] in col_list:
                if wall[x][y + 1] not in bo_list and wall[x + 1][y + 1] not in col_list:
                    continue
            if wall[x][y - 1] == 1 or wall[x][y + 1] == 1: # 제거하려는 보 양 옆에 양쪽으로 연결되지 않은 보가 남게 되는 경우
                continue
            wall[x][y] == 3 if wall[x][y] != 2 else 0
            
    # 작업 완료 후
    for i in range(n + 1):
        for j in range(n + 1):
            x = j
            y = n - i
            if wall[i][j] == 0: # 기둥만 설치되어 있는 경우     
                answer.append([x, y, 0])
            elif wall[i][j] == 1: # 보만 설치되어 있는 경우
                answer.append([x, y, 1])
            elif wall[i][j] == 2: # 기둥과 보 모두 설치되어 있는 경우
                answer.append([x, y, 0], [x, y, 1])
            else:
                continue
    
    answer.sort()
        
    return answer
```

```python
def delete_bo(x, y, wall):
    bo_list = [1, 2]
    col_list = [0, 2]
    if wall[x][y] in col_list:
        if wall[x + 1][y] not in col_list and wall[x][y - 1] not in bo_list: # 제거하려는 보 위에 기둥만 남게 되는 경우
            return False
    if wall[x][y + 1] in col_list:
        if wall[x][y + 1] not in bo_list and wall[x + 1][y + 1] not in col_list:
            return False
    if wall[x][y - 1] == 1 or wall[x][y + 1] == 1: # 제거하려는 보 양 옆에 양쪽으로 연결되지 않은 보가 남게 되는 경우
        return False
    return True

def delete_col(x, y, wall):
    bo_list = [1, 2]
    col_list = [0, 2]
    # 제거하려는 기둥 위에 기둥만 있을 경우 
    if wall[x - 1][y] == 0 and wall[x - 1][y - 1] not in bo_list:
        return False
    # 제거하려는 기둥 위(오른쪽)에 보가 있을 경우    
    if wall[x - 1][y] in bo_list:
        if not (wall[x - 1][y - 1] in bo_list and wall[x - 1][y + 1] in bo_list):
            return False
    # 제거하려는 기둥 위(왼쪽)에 보가 있을 경우 
    if wall[x - 1][y - 1] in bo_list:
        if not (wall[x - 1][y - 2] in bo_list and wall[x - 1][y] in bo_list):
            return False
    return True

def solution(n, build_frame):
    answer = []
    # build_frame의 원소 [x, y, a, b]
    # x, y는 기둥 또는 보를 설치할 교차점의 좌표
    # a는 구조물의 종류 - 기둥은 0, 보는 1
    # b는 작업의 종류 - 삭제 0, 설치 1
    
    wall = [[3] * (3 * n + 3) for _ in range(3 * n + 3)] # 3n + 3인 2차원 행렬 초기화 (3로 채움)
    
    # 작업 수만큼 반복
    for i in range(len(build_frame)):
        # 벽면 배열 인덱스에 맞도록 주어진 위치 값 조정
        x = n - build_frame[i][1] + (n + 1)
        y = build_frame[i][0] + (n + 1)
        bo_list = [1, 2]
        col_list = [0, 2]
        
        build_type = build_frame[i][2] # 기둥인지 보인지 확인 
        build = True if build_frame[i][3] == 1 else False # 설치할지 삭제할지 확인

        # 기둥을 설치하는 경우
        if build_type == 0 and build:
            if x == 2 * n + 1 or wall[x][y - 1] in bo_list or wall[x + 1][y] in col_list: 
                wall[x][y] = 0 if wall[x][y] != 1 else 2 # 기둥과 보가 모두 설치되는 칸인 경우 2 할당
        
        # 보를 설치하는 경우
        elif build_type == 1 and build:
            if wall[x + 1][y] in col_list or wall[x + 1][y + 1] in col_list or (wall[x][y + 1] in bo_list and wall[x][y - 1] in bo_list): 
                wall[x][y] = 1 if wall[x][y] != 0 else 2 # 기둥과 보가 모두 설치되는 칸인 경우 2 할당
        
        # 기둥을 제거하는 경우
        elif build_type == 0 and not build: 
            if delete_col(x, y, wall): wall[x][y] = 3 if wall[x][y] != 2 else 1
            
        # 보를 제거하는 경우
        else:
            if delete_bo(x, y, wall): wall[x][y] == 3 if wall[x][y] != 2 else 0
            
    # 작업 완료 후
    for i in range(n + 1, 2 * n + 2):
        for j in range(n + 1, 2 * n + 2):
            x = j - n - 1
            y = 2 * n + 1 - i
            if wall[i][j] == 0: # 기둥만 설치되어 있는 경우     
                answer.append([x, y, 0])
            elif wall[i][j] == 1: # 보만 설치되어 있는 경우
                answer.append([x, y, 1])
            elif wall[i][j] == 2: # 기둥과 보 모두 설치되어 있는 경우
                answer.append([x, y, 0], [x, y, 1])
            else:
                pass
    
    answer.sort()
        
    return answer
```

### 정답 코드
- 시간 제한이 넉넉한 편이므로 설치 및 삭제 연산을 요구할 때마다 일일이 '전체 구조물을 확인하며' 규칙 롹인
- possible() 메서드를 만들어 현재의 구조물이 정상인지 체크

```python
# 현재 설치된 구조물이 '가능한' 구조물인지 확인하는 함수
def possible(answer):
    for x, y, stuff in answer:
        if stuff == 0: # 설치된 것이 '기둥'인 경우
            # 바닥 위 혹은 보의 한쪽 끝부분 위 혹은 다른 기둥 위라면 정상
            if y == 0 or [x - 1, y, 1] in answer or [x, y, 1] in answer or [x, y - 1, 0] in answer:
                continue
            return False # 아니라면 False 반환
        elif stuff == 1: # 설치된 것이 '보' 인 경우
            # 한쪽 끝부분이 기둥 위 혹은 양쪽 끝부분이 다른 보와 동시에 연결 이라면 정상
            if [x, y - 1, 0] in answer or [x + 1, y - 1, 0] in answer or ([x - 1, y, 1] in answer and [x + 1, y, 1] in answer):
                continue
            return False # 아니라면 False 반환
    return True

def solution(n, build_frame):
    answer = []
    for frame in build_frame: # 작업(frame)의 개수는 최대 1000개
        x, y, stuff, operate = frame
        if operate == 0: # 삭제하는 경우
            answer.remove([x, y, stuff]) # 일단 삭제를 해본 뒤에
            if not possible(answer): # 가능한 구조물인지 확인
                answer.append([x, y, stuff]) # 가능한 구조물이 아니라면 다시 설치
        if operate == 1:
            answer.append([x, y, stuff]) # 일단 설치를 해본 뒤에
            if not possible(answer): # 가능한 구조물인지 확인
                answer.remove([x, y, stuff]) # 가능한 구조물이 아니라면 다시 제거
    return sorted(answer) # 정렬된 결과를 반환
```

## 기출문제 6 치킨 배달
도시에 있는 치킨집 중에서 최대 M개를 고르고 나머지 치킨집을 모두 폐업시킬 때 어떻게 고르면 도시의 치킨 거리가 가장 작게 되는가?

### 내 풀이

```python
# 입력 조건
# 첫째 줄에 n(도시 넓이), m(최대 고를 치킨집 수)이 주어짐
n, m = map(int, input().split())
# 둘째 줄부터 도시의 정보가 주어짐 0 - 빈칸, 1 - 집, 2 - 치킨집
city_map = []
for _ in range(n):
    city_map.append(list(map(int, input().split())))
    
chickens = []
houses = []
# 치킨집 및 집 인덱스 저장
for i in range(n):
    for j in range(n):
        if city_map[i][j] == 2: chickens.append((i, j))
        elif city_map[i][j] == 1: houses.append((i, j))
        else: pass
            
# 치킨집 중에서 m개를 고르는 조합 구하기 
from itertools import combinations
combi = list(combinations(chickens, m))

result = 999999999
# 모든 조합의 경우에서 도시의 치킨 거리 계산하기
for i in range(len(combi)):
    distance_sum = 0 # 도시의 치킨 거리 
    for j in range(len(houses)): # 각 집의 관점에서 가장 가까운 치킨집까지의 거리 계산
        min_distance = 999999999
        for k in range(m): # 모든 치킨집과의 거리를 각각 계산해서 가장 가까운 거리만 저장
            temp = combi[i][k] # 조합으로 선택된 치킨집 리스트 
            distance = abs(houses[j][0] - temp[0]) + abs(houses[j][1] - temp[1])
            if distance <= min_distance: min_distance = distance
        distance_sum += min_distance
    if distance_sum <= result: result = distance_sum
        
print(result)
```

### 정답 코드

```python
from itertools import combinations

n, m = map(int, input().split())
chicken, house = [], []

for r in range(n):
    data = list(map(int, input().spit()))
    for c in range(n):
        if data[c] == 1:
            house.append((r, c)) # 일반 집
        elif data[c] == 2:
            chicken.append((r, c)) # 치킨집
            
# 모든 치킨집 중에서 m개의 치킨집을 뽑는 조합 계산
candidates = list(combinations(chicken, m))

# 치킨 거리의 합을 계산하는 함수
def get_sum(candidate):
    result = 0
    # 모든 집에 대하여
    for hx, hy in house:
        # 가장 가까운 치킨집을 찾기
        temp = 1e9
        for cx, cy in candidate:
            temp = min(temp, abs(hx - cx) + abs(hy - cy))
        # 가장 가까운 치킨집까지의 거리를 더하기
        result += temp
    return result

# 치킨 거리의 합의 최소를 찾아 출력
result = 1e9
for candidate in candidates:
    result = min(result, get_sum(candidate))
    
print(result)
```

## 기출문제 7 외벽 점검
외벽의 길이 n, 취약 지점의 위치가 담긴 배열 weak, 각 친구가 1시간 동안 이동할 수 있는 거리가 담긴 dist가 매개변수로 주어질 때, 취약 지점을 점검하기 위해 보내야 하는 친구 수의 최솟값을 return 하도록 solution 함수를 완성하라  
친구들을 모두 투입해도 취약 지점을 전부 점검할 수 없는 경우 -1을 return

### 정답 코드

```python
from itertools import permutations
# 순열

def solution(n, weak, dist):
    # 길이를 2배로 늘려서 '원형'을 일자 형태로 변형
    length = len(weak)
    for i in range(length):
        weak_append(weak[i] + n)
        answer = len(dist) + 1 # 투입할 친구 수의 최솟값을 찾아야 하므로 len(dist) + 1로 초기화
        # 0부터 length - 1 까지의 위치를 각각 시작점으로 설정
        for start in range(length):
            # 친구를 나열하는 모든 경우의수 각각에 대하여 확인
            for friends in list(permutations(dist, len(dist))):
                count = 1 # 투입할 친구의 수
                # 해당 친구가 점검할 수 있는 마지막 위치
                position = weak[start] + friends[count - 1]
                # 시작점부터 모든 취약 지점을 확인
                for index in range(start, start + length)
                # 점검할 수 있는 위치를 벗어나는 경우
                if position < weak[index]:
                    count += 1 # 새로운 친구를 도입
                    if count > len(dist): # 더 투입이 불가능하다면 종료
                        break
                    position = weak[index] + friends[count - 1]
                answer = min(answer, count) # 최솟값 계산
    if answer > len(dist):
        return -1
    return answer
```
