## 예제 15 특정 거리의 도시 찾기
- 1~N번까지의 도시와 M개의 단방향 도로가 존재 (모든 도로의 거리는 1)
- 특정한 도시 X로부터 출발하여 도달할 수 있는 모든 도시 중에서, 최단 거리가 정확히 K인 모든 도시의 번호를 출력하라

### 정답 코드

```python
# 입력 조건
# 첫째 줄에 도시의 개수 N, 도로의 개수 M, 거리 정보 K, 출발 도시 번호 X가 주어짐
n, m, k, x = map(int, input().split())
# 둘째 줄부터 M개 줄에 걸쳐서 두 개 자연수 A, B가 주어짐
# (A번 도시에서 B번 도시로 이동하는 단방향 도로가 존재한다는 의미)
graph = [[] for _ in range(n + 1)]
for _ in range(m):
    a, b = map(int, input().split())
    graph[a].append(b)
    
# BFS 문제 -> 큐 이용
from collections import deque

# 모든 도시에 대한 최단 거리 초기화
distance = [-1] * (n + 1)
distance[x] = 0 # 출발 도시까지의 거리는 0으로 설정

# 너비 우선 탐색 BFS 수행
q = deque([x])
while q:
    now = q.popleft()
    # 현재 도시에서 이동할 수 있는 모든 도시 확인
    for next_node in graph[now]:
        # 아직 방문하지 않은 도시라면
        if distance[next_node] == -1:
            # 최단 거리 갱신
            distance[next_node] = distance[now] + 1
            q.append(next_node)

# 최단 거리가 K인 모든 도시의 번호를 오름차순으로 출력
check = False
for i in range(1, n + 1):
    if distance[i] == k:
        print(i)
        check = True
        
# 만약 최단 거리인 도시가 없으면
if check == False:
    print(-1)
```

## 예제 16 연구소
연구소의 지도가 주어졌을 때 벽을 3개 세운 뒤 바이러스가 퍼질 수 없는 안전 영역 크기의 최댓값을 구하라 

### 내 코드 
틀렸는데 뭐가 잘못된건지 모르겠음 

```python
# 입력 조건
# 첫째 줄에 지도의 세로 크기 n과 가로 크기 m이 주어짐
n, m = map(int, input().split())
# 둘째 줄부터 n개 줄에 지도의 모양이 주어짐 (0:빈칸, 1:벽, 2:바이러스)
lab = []
for _ in range(n):
    lab.append(list(map(int, input().split())))
    
empty = [] # 벽을 세울 수 있는 빈 공간
virus = [] # 바이러스가 있는 공간
for i in range(n):
    for j in range(m):
        if lab[i][j] == 0:
            empty.append((i, j))
        if lab[i][j] == 2:
            virus.append((i, j))

# 빈 칸에서 3개를 골라 벽을 세우는 모든 경우를 고려하자 (조합)
from itertools import combinations
candidates = list(combinations(empty, 3))

# dfs로 특정 노드(바이러스)를 방문한 뒤에 연결된 모든 노드들도 방문
def dfs(x, y):
    # 주어진 범위를 벗어나는 경우 즉시 종료
    if x <= -1 or x >= n or y <= -1 or y >= m:
        return False
    # 벽을 만난 경우 종료
    if temp[x][y] == 1:
        return False
    # 현재 노드를 아직 방문하지 않았다면
    if temp[x][y] == 0:
        # 해당 노드 방문 처리 (바이러스 퍼짐)
        temp[x][y] = 2
        # 상, 하, 좌, 우의 위치도 모두 재귀적으로 호출
        dfs(x - 1, y)
        dfs(x, y - 1)
        dfs(x + 1, y)
        dfs(x, y + 1)
        return True
    return False

# 상, 하, 좌, 우로 바이러스가 퍼짐
# dx = [-1, 1, 0, 0]
# dy = [0, 0, -1, 1]

result = 0 # 최댓값을 0으로 초기화

for i in range(len(candidates)):
    count = 0
    temp = lab.copy()
    # 벽 세우기
    for j in range(3):
        x = candidates[i][j][0]
        y = candidates[i][j][1]
        temp[x][y] = 1
        
    # 벽 세운 후 모든 바이러스에 대해 dfs 수행
    for v in virus:
        dfs(v[0], v[1])         
        
    # 안전영역 크기 계산하기 - 빈 곳이 채워졌는지 확인하면 됨
    for i in range(n):
        for j in range(m):
            if temp[i][j] == 0:
                count += 1
    
    if count >= result: result = count 
        
print(result)
```

### 정답 코드

```python
n, m = map(int, input().split())
data = [] # 초기 맵 리스트
temp = [[0] * m for _ in range(n)] # 벽을 설치한 뒤의 맵 리스트

for _ in range(n):
    data.append(list(map(int, input().split())))

# 4가지 이동 방향에 대한 리스트
dx = [-1, 0, 1, 0]
dy = [0, 1, 0, -1]

result = 0

# 깊이 우선 탐색(DFS)을 이용해 각 바이러스가 사방으로 퍼지도록 하기
def virus(x, y):
    for i in range(4):
        nx = x + dx[i]
        ny = y + dy[i]
        # 상, 하, 좌, 우 중에서 바이러스가 퍼질 수 있는 경우
        if nx >= 0 and nx < n and ny >= 0 and ny < m:
            if temp[nx][ny] == 0:
                # 해당 위치에 바이러스 배치하고, 다시 재귀적으로 수행
                temp[nx][ny] = 2
                virus(nx, ny)

# 현재 맵에서 안전 영역의 크기 계산하는 메서드
def get_score():
    score = 0
    for i in range(n):
        for j in range(m):
            if temp[i][j] == 0:
                score += 1
    return score

# 깊이 우선 탐색(DFS)을 이용해 울타리를 설치하면서, 매 번 안전 영역의 크기 계산
def dfs(count):
    global result
    # 울타리가 3개 설치된 경우
    if count == 3:
        for i in range(n):
            for j in range(m):
                temp[i][j] = data[i][j]
        # 각 바이러스의 위치에서 전파 진행
        for i in range(n):
            for j in range(m):
                if temp[i][j] == 2:
                    virus(i, j)
        # 안전 영역의 최대값 계산
        result = max(result, get_score())
        return
    # 빈 공간에 울타리를 설치
    for i in range(n):
        for j in range(m):
            if data[i][j] == 0:
                data[i][j] = 1
                count += 1
                dfs(count)
                data[i][j] = 0
                count -= 1

dfs(0)
print(result)
```

## 예제 17 경쟁적 전염
N X N 크기의 시험관에서 바이러스가 낮은 번호부터 매 초마다 증식할 때, S초가 지난 후 (X, Y)에 존재하는 바이러스의 종류를 출력하라

### 내 코드 
틀려씀

```python
# 입력 조건
# 첫째 줄에 자연수 N, K가 주어짐
n, k = map(int, input().split())
# 둘째 줄부터 N개 줄에 걸쳐 시험관 정보가 주어짐
# 해당 위치에 존재하는 바이러스의 번호가 주어짐, 비어있는 경우 0
lab = []
for _ in range(n):
    lab.append(list(map(int, input().split())))
print(lab)

# 초기 바이러스 위치 저장
virus = [[] for _ in range(k + 1)]
for i in range(n):
    for j in range(n):
        if lab[i][j] != 0:
            virus[lab[i][j]].append((i, j))

# N + 2번째 줄에는 S, X, Y가 주어짐
s, x, y = map(int, input().split())
print(s, '초 후에', x, '와', y, '에 위치한 바이러스의 번호를 출력하라')

# 4가지 이동 방향에 대한 리스트
dx = [-1, 0, 1, 0]
dy = [0, 1, 0, -1]

def dfs(i, j, virus_num):
    for num in range(4):
        ni = i + dx[num]
        nj = j + dy[num]
        # 바이러스가 퍼질 수 있는 공간인지 확인
        if ni >= 0 and ni < n and nj >= 0 and nj < n:
            if lab[ni][nj] == 0:
                # 해당 위치에 바이러스 배치하고, 다시 재귀적으로 수행
                lab[ni][nj] = virus_num
                virus[virus_num].append((i, j))
            
# S초가 될 때까지 전염
result = 0
count = 0
while True:
    for i in range(1, k + 1):
        for v in range(len(virus[i])):
            print(i, '번 바이러스가 전염된 칸 개수: ', len(virus[i]))
            a = virus[i][v][0]
            b = virus[i][v][1]
            dfs(a, b, i)
    count += 1
    
    print(count, '초 뒤에 시험관 모습')
    print(lab)
    
    if count == s:
        break
    
# S초 뒤에 (X, Y)에 존재하는 바이러스 종류 출력
result = lab[x - 1][y - 1]
print(result)
```

### 정답 코드
- 너비 우선 탐색
- 바이러스가 낮은 번호부터 증식하므로 초기에 큐에 원소를 삽입할 때 낮은 바이러스의 번호부터 삽입

```python
from collections import deque

n, k = map(int, input().split())

graph = [] # 전체 보드 정보를 담는 리스트
data = [] # 바이러스에 대한 정보를 담는 리스트

for i in range(n):
    # 보드 정보를 한 줄 단위로 입력
    graph.append(list(map(int, input().split())))
    for j in range(n):
        # 해당 위치에 바이러스가 존재하는 경우
        if graph[i][j] != 0:
            # (바이러스 종류, 시간, 위치 X, 위치 Y) 삽입
            data.append((graph[i][j], 0, i, j))

# 정렬 이후에 큐로 옮기기 (낮은 번호의 바이러스가 먼저 증식하므로)
data.sort()
q = deque(data)
 
target_s, target_x, target_y = map(int, input().split())
 
# 바이러스가 퍼져나갈 수 있는 4가지의 위치
dx = [-1, 0, 1, 0]
dy = [0, 1, 0, -1]

# 너비 우선 탐색(BFS) 진행
while q:
    virus, s, x, y = q.popleft()
    # 정확히 s초가 지나거나, 큐가 빌 때까지 반복
    if s == target_s:
        break
    # 현재 노드에서 주변 4가지 위치를 각각 확인
    for i in range(4):
        nx = x + dx[i]
        ny = y + dy[i]
        # 해당 위치로 이동할 수 있는 경우
        if 0 <= nx and nx < n and 0 <= ny and ny < n:
            # 아직 방문하지 않은 위치라면, 그 위치에 바이러스 넣기
            if graph[nx][ny] == 0:
                graph[nx][ny] = virus
                q.append((virus, s + 1, nx, ny))

print(graph[target_x - 1][target_y - 1])
```

## 예제 18 괄호 변환
괄호를 올바른 순서대로 배치하기

### 내 풀이 + 정답 코드 짬뽕 (두번째 함수)

```python
# 균형잡힌 괄호 문자열 u를 구하기 위해 인덱스를 반환하는 함수
def balanced(p):
    count = 0 # 왼쪽 괄호를 센다
    for i in range(len(p)):
        if p[i] == '(':
            count += 1
        else:
            count -= 1
        if count == 0:
            return i
        
# 올바른 괄호 문자열인지 판별하는 함수
def check(p):
    count = 0 # 왼쪽 괄호를 센다
    for i in range(len(p)):
        if p[i] == '(':
            count += 1
        else:
            if count == 0:
                return False
            count -= 1
    return True
    
def solution(p):
    answer = ''
    if p == '':
        return answer
    u = p[:balanced(p) + 1]
    v = p[balanced(p) + 1:]

    if check(u):
        answer = u + solution(v)
    
    else:
        answer = '('
        answer += solution(v) + ')'
        u = list(u[1:-1])
        for i in range(len(u)):
            if u[i] == '(': u[i] = ')'
            else: u[i] = '('
        answer += ''.join(u)
            
    return answer

solution('()))((()')
```

## 예제 19 연산자 끼워 넣기
N개의 수로 이루어진 수열에 N - 1개 연산자를 끼워 넣을 때 만들 수 있는 식의 결과가 최대인 것과 최소인 것을 구하라
- 나눗셈은 정수 나눗셈으로 몫만 취함
- 음수를 양수로 나눌 때는 양수로 바꾼 뒤 몫을 취하고 그 몫을 음수로 바꾼다

### 내 풀이
- 중복 순열로 풀 수 있음

```python
# 입력 조건
# 첫째 줄에 수의 개수 n이 주어짐
n = int(input())
# 둘째 줄에 수열이 주어짐
num_list = list(map(int, input().split()))
# 셋째 줄에 합이 n-1개인 4개 정수가 주어짐, 차례대로 +, -, x, /의 개수
add, sub, mul, div = map(int, input().split())
data = ['add' for _ in range(add)] + ['sub' for _ in range(sub)]
data += ['mul' for _ in range(mul)] + ['div' for _ in range(div)]
    
# 순열 
from itertools import permutations
candidates = list(permutations(data, n - 1))

target_max = -1e9
target_min = 1e9

# 연산 함수
def cal(a, b, string):
    if string == 'add':
        return a + b
    if string == 'sub':
        return a - b
    if string == 'mul':
        return a * b
    if string == 'div':
        if a < 0 and b > 0:
            return (a * (-1) // b) * (-1)
        return a // b

# 모든 경우에 대해 계산해보기
for c in range(len(candidates)):
    temp = num_list.copy()
    for i in range(n - 1):
        a = temp[i]
        b = temp[i + 1]
        temp[i + 1] = cal(a, b, candidates[c][i])
        
    target_max = max(target_max, temp[n - 1])
    target_min = min(target_min, temp[n - 1])
            
print(target_max)
print(target_min)
```

### 정답 코드
- DFS 이용하여 풀기
- 코드를 봐도 잘 이해가 안된다 -> 재귀함수 이해부터 확실히 해야 할 듯

```python
n = int(input())
# 연산을 수행하고자 하는 수 리스트
data = list(map(int, input().split()))
# 더하기, 빼기, 곱하기, 나누기 연산자 개수
add, sub, mul, div = map(int, input().split())

# 최솟값과 최댓값 초기화
min_value = 1e9
max_value = -1e9

# 깊이 우선 탐색(DFS) 메서드
def dfs(i, now):
    global min_value, max_value, add, sub, mul, div
    # 모든 연산자를 다 사용한 경우, 최솟값과 최댓값 업데이트
    if i == n:
        min_value = min(min_value, now)
        max_value = max(max_value, now)
    else:
        # 각 연산자에 대하여 재귀적으로 수행
        if add > 0:
            add -= 1
            dfs(i + 1, now + data[i])
            add += 1
        if sub > 0:
            sub -= 1
            dfs(i + 1, now - data[i])
            sub += 1
        if mul > 0:
            mul -= 1
            dfs(i + 1, now * data[i])
            mul += 1
        if div > 0:
            div -= 1
            dfs(i + 1, int(now / data[i])) # 나눌 때는 나머지를 제거
            div += 1

dfs(1, data[0])

print(max_value)
print(min_value)
```

## 예제 20 감시 피하기
N X N 크기의 복도에 장애물을 3개 설치할 때 선생님의 감시로부터 모든 학생이 피할 수 있는지 계산하라

### 정답코드

```python
# 입력 조건
# 첫째 줄에 자연수 N이 주어짐
n = int(input())
# 둘째 줄에 복도의 정보가 주어짐
# 각 행의 해당 위치에 학생이 있다면 S, 선생님이 있다면 T 아무것도 존재하지 않는다면 X
board, teachers, spaces = [], [], []
for i in range(n):
    board.append(list(input().split()))
    for j in range(n):
        if board[i][j] == 'T':
            teachers.append((i, j))
        if board[i][j] == 'X':
            spaces.append((i, j))
            
# 특정 방향으로 감시를 진행(학생 발견 시 True 반환)
def watch(x, y, direction):
    # 왼쪽 방향으로 감시
    if direction == 0:
        while y >= 0:
            if board[i][j] == 'S': # 학생이 있는 경우
                return True
            if board[i][j] == 'O': # 장애물이 있는 경우
                return False
            y -= 1
    # 오른쪽 방향으로 감시
    if direction == 1:
        while y < n:
            if board[i][j] == 'S': # 학생이 있는 경우
                return True
            if board[i][j] == 'O': # 장애물이 있는 경우
                return False
            y += 1
    # 위쪽 방향으로 감시
    if direction == 2:
        while x >= 0:
            if board[i][j] == 'S': # 학생이 있는 경우
                return True
            if board[i][j] == 'O': # 장애물이 있는 경우
                return False
            x -= 1
    # 아래쪽 방향으로 감시
    if direction == 3:
        while x < n:
            if board[i][j] == 'S': # 학생이 있는 경우
                return True
            if board[i][j] == 'O': # 장애물이 있는 경우
                return False
            x += 1
    return False

# 장애물 설치 이후 한 명이라도 학생이 감시되는지 검사
def process():
    for x, y in teachers:
        # 네 가지 방향 모두 검사
        for i in range(4):
            if watch(x, y, i):
                return True # 학생이 감지된 경우
    return False

result = False # 학생이 한 명도 감지되지 않도록 장애물을 설치할 수 있는지의 여부

# 장애물을 설치할 모든 경우의 수(조합)에 대해 검사
from itertools import combinations
for data in combinations(spaces, 3):
    # 장애물 설치
    for x, y in data:
        board[x][y] = 'O'
    # 학생이 한 명도 감지되지 않는 경우
    if not process():
        result = True
        break
    # 그렇지 않은 경우
    # 설치된 장애물 다시 없애기
    for x, y in data:
        board[x][y] = 'X'
        
if result: print('YES')
else: print("NO")
```

## 예제 21 인구 이동
국경선을 공유하는 두 나라의 인구 차이가 L명 이상 R명 이하라면 국경선을 열고 연랍을 이룬다.  
연합을 이루는 각 칸의 인구수는 (연합의 인구수) / (연합을 이루고 있는 칸의 개수)가 된다. (소수점 버림)  
각 나라의 인구수가 주어졌을 때, 인구 이동이 몇 번 발생하는지 구하라.

### 정답 코드

```python
# 입력 조건
# 땅의 크기(N), L, R값을 입력받기
n, l, r = map(int, input().split())

# 전체 나라의 정보(N X N) 입력받기
graph = []
for _ in range(n):
    graph.append(list(map(int, input().split())))
    
dx = [-1, 0, 1, 0]
dy = [0, -1, -, 1]


# 특정 위치에서 출발하여 모든 연합을 체크한 뒤 데이터 갱신
def process(x, y, index):
    #(x, y)의 위치와 연결된 나라(연합) 정보를 담는 리스트
    united = []
    united.append((x, y))
    # 너비 우선 탐색(BFS)을 위한 큐 자료구조 정의
    q = deque()
    q.append((x, y))
    union[x][y] = index # 현재 연합의 번호 할당
    summary = graph[x][y] # 현재 연합의 전체 인구 수
    count = 1 # 현재 연합의 국가 수
    # 큐가 빌 때까지 반복(BFS)
    while q:
        x, y = q.popleft()
        # 현재 위치에서 4가지 방향을 확인하며
        for i in range(4):
            nx = x + dx[i]
            ny = y + dy[i]
            # 바로 옆에 있는 나라를 확인하며
            if 0 <= nx < n and 0 <= ny < n and union[nx][ny] == -1:
                # 옆에 있는 나라와 인구 차이가 L명 이상, R명 이하라면
                if l <= abs(graph[nx][ny] - graph[x][y]) <= r:
                    q.append((nx, ny))
                    # 연합에 추가하기 
                    union[nx][ny] = index
                    summary += graph[nx][ny]
                    count += 1
                    united.append((nx, ny))
    # 연합 국가끼리 인구 분배
    for i, j in united:
        graph[i][j] = summary // count

total_count = 0

# 더 이상 인구 이동을 할 수 없을 때까지 반복
while True:
    union = [[-1] * n for _ in range(n)]
    index = 0
    for i in range(n):
        for j in range(n):
            if union[i][j] == -1: # 해당 나라가 아직 처리되지 않았다면
                process(i, j, index)
                index += 1
    # 모든 인구 이동이 끝난 경우
    if index == n * n:
        break
    total_count += 1
    
# 인구 이동 횟수 출력
print(total_count)
```
