## 1. 크레인 인형뽑기 게임
- 단순 구현 문제

```python
def solution(board, moves):
    answer = 0
    basket = []
    
    for m in moves:
        j = m - 1
        for i in range(len(board)):
            if board[i][j] != 0:
                if len(basket) >= 1 and basket[-1] == board[i][j]:
                    answer += 2
                    basket.pop()
                else:
                    basket.append(board[i][j])
                board[i][j] = 0
                break
            
    return answer
```

## 2. 두 개 뽑아서 더하기


```python
def solution(numbers):
    answer = []
    for i in range(len(numbers) - 1):
        for j in range(i + 1, len(numbers)):
            answer.append(numbers[i] + numbers[j])
            
    answer = list(set(answer))
    answer.sort()
    return answer
```

### 내 코드

```python
import re

def solution(new_id):
    answer = ''
    # step1
    new_id = new_id.lower()
    # step2
    new_id = re.sub('[\~\!\@\#\$\%\^\&\*\(\)\=\+\[\{\]\}\:\?\,\<\>\/]', '', new_id)
    # step3
    new_id = re.sub('[.]{2,1000}', '.', new_id)
    # step4
    new_id = new_id.strip('.')
    # step5
    if new_id == '':
        new_id = "a"
    # step6
    if len(new_id) >= 16:
        new_id = new_id[:15]
        new_id = new_id.strip('.')
    # step7
    if len(new_id) <= 2:
        while len(new_id) <= 2:
            new_id += new_id[-1]
    answer = new_id
    return answer
```

## 7. 두 정수 사이의 합


### 내 풀이
- abs는 잘 썼는데 더 일반화시킬 수 있었다

```python
def solution(a, b):
    if a == b:
        return a
    # 더하는 숫자가 짝수개인 경우
    if (abs(a - b) + 1) % 2 == 0:
        return (a + b) * (abs(a - b) + 1) / 2
    # 더하는 숫자가 홀수개인 경우
    if (abs(a - b) + 1) % 2 != 0:
        return (a + b) * abs(a - b) / 2 + (a + b) / 2
```

## 9. 문자열 내 p와 y의 개수


### 내 코드
- 단순 포문

```python
def solution(s):
    s = s.lower()
    p_count, y_count = 0, 0
    for x in list(s):
        if x == 'p': p_count += 1
        if x == 'y': y_count += 1
    return True if p_count == y_count else False
```

## 13. 약수의 합


### 상위 코드

```python
def solution(num):
    # num / 2 의 수들만 검사하면 성능 약 2배 향상잼
    # 자기 자신은 무조건 약수로 가짐
    # 자기 자신보다 하나 더 작은 약수는 2로 나누어 떨어지는 경우 발생
    return num + sum([i for i in range(1, (num // 2) + 1) if num % i == 0])
```

## 15. 소수 찾기


### 내 코드
- 에라토스테네스의 체
- 숙지가 안된 것 같다 

```python
def solution(n):
    arr = [True for i in range(n + 1)]
    arr[0], arr[1] = False, False # 1은 소수가 아니므로
    for i in range(2, int(n ** (1/2)) + 1):
        if arr[i] == True:
            j = 2
            while i * j <= n:
                arr[i * j] = False
                j += 1
            
    return arr.count(True)
```

### 상위 코드
- set 차집합 연산으로 구현

```python
def solution(n):
    num = set(range(2, n + 1))

    for i in range(2, n + 1):
        if i in num:
            num -= set(range(2 * i, n + 1,i))
            
    return len(num)
```

## 16. [카카오 인턴십] 키패드 누르기


### 내 코드
- 번호의 위치를 딕셔너리에 저장

```python
def solution(numbers, hand):
    left_loc, right_loc = ['*'], ['#']
    keypad = {1:[0, 0], 2:[0, 1], 3:[0, 2], 4:[1, 0], 5:[1, 1], 6:[1, 2], 7:[2, 0], 8:[2, 1], 9:[2, 2], '*':[3, 0], 0:[3, 1], '#':[3, 2]}
    # [[1, 2, 3], [4, 5, 6], [7, 8, 9], ['*', 0, '#']]
    result = ""
    for x in numbers:
        if x in [1, 4, 7]:
            result += 'L'
            left_loc.append(x)
        elif x in [3, 6, 9]:
            result += 'R'
            right_loc.append(x)
        else: # 2, 5, 8, 0을 눌러야 하는 경우
            # 왼손 거리와 오른손 거리 비교
            left_distance = abs(keypad[left_loc[-1]][0] - keypad[x][0]) + abs(keypad[left_loc[-1]][1] - keypad[x][1])
            right_distance = abs(keypad[right_loc[-1]][0] - keypad[x][0]) + abs(keypad[right_loc[-1]][1] - keypad[x][1])
            if left_distance > right_distance:
                result += 'R'
                right_loc.append(x)
            elif left_distance < right_distance:
                result += 'L'
                left_loc.append(x)
            else:
                if hand == 'right':
                    result += 'R'
                    right_loc.append(x)
                else:
                    result += 'L'
                    left_loc.append(x)
    return result
```

## 17. 소수 만들기


### 내 코드

```python
import itertools

def solution(nums):
    # 3000 이하의 소수를 모두 구하고 시작
    arr = [True] * 3001
    arr[0], arr[1] = False, False
    for i in range(2, int(3000 ** (1/2)) + 1):
        if arr[i] == True:
            j = 2
            while i * j <= 3000:
                arr[i * j] = False
                j += 1
                
    # 주어진 숫자 중 3개를 뽑아 더하기
    answer = 0
    candidates = list(itertools.combinations(nums, 3))
    for x in candidates:
        if arr[sum(x)]:
            answer += 1
    return answer
```

## 18. 비밀지도


### 내 코드
- bin()

```python
def solution(n, arr1, arr2):
    # 암호화된 배열을 이진수로 변환하기
    map1, map2 = [], []
    for i, j in zip(arr1, arr2):
        new_i = bin(i)[2:] if len(bin(i)[2:]) == n else '0' * (n - len(bin(i)[2:])) + bin(i)[2:]
        map1.append(new_i)
        new_j = bin(j)[2:] if len(bin(j)[2:]) == n else '0' * (n - len(bin(j)[2:])) + bin(j)[2:]
        map2.append(new_j)
        
    # 비밀지도 복원
    result = []
    for a, b in zip(map1, map2):
        row = ''
        for i in range(n):
            if a[i] == '0' and b[i] == '0':
                row += ' '
            elif a[i] == '1' or b[i] == '1':
                row += '#'
        result.append(row)

    return result
```

## 19. 다트게임


### 내 코드

```python
def solution(dartResult):
    multi = {'S':1, 'D':2, 'T':3}
    arr = []
    score = ''
    for i in range(len(dartResult)):
        if dartResult[i].isdigit():
            score += dartResult[i]
        elif dartResult[i] in ['S', 'D', 'T']:
            score += dartResult[i]
            arr.append(score)
            score = ''
        else:
            arr.append(dartResult[i])
            score = ''
    
    answer = []
    for x in arr:
        if x == '*':
            if len(answer) >= 2:
                answer[-2] *= 2
            answer[-1] *= 2
        elif x == '#':
            answer[-1] *= (-1)
        else:
            answer.append(int(x[:-1]) ** multi[x[-1]])
    return sum(answer)
```

### 상위 코드
- 역시 문자열은 정규표현식이다 잘 알아두자

```python
import re

def solution(dartResult):
    bonus = {'S' : 1, 'D' : 2, 'T' : 3}
    option = {'' : 1, '*' : 2, '#' : -1}
    p = re.compile('(\d+)([SDT])([*#]?)')
    dart = p.findall(dartResult)
    for i in range(len(dart)):
        if dart[i][2] == '*' and i > 0:
            dart[i-1] *= 2
        dart[i] = int(dart[i][0]) ** bonus[dart[i][1]] * option[dart[i][2]]

    answer = sum(dart)
    return answer
```

## 20. 최대공약수와 최소공배수


### 내 코드

```python
def solution(n, m):
    div_n, div_m = [1, n], [1, m]
    
    # 각 수의 약수 구하기
    div_n += [i for i in range(2, int(n / 2) + 1) if n % i == 0]
    div_m += [i for i in range(2, int(m / 2) + 1) if m % i == 0]
    # 각 수의 배수 구하기
    mul_n = [i for i in range(n, n * m + 1, n)]
    mul_m = [i for i in range(m, n * m + 1, m)]
    
    # 최대공약수 구하기
    a = max(set(div_n) & set(div_m))
    # 최소공배수 구하기
    b = min(set(mul_n) & set(mul_m))
    
    return [a, b]
```

## 21. 시저 암호


### 내 코드
- 문자 -> 아스키코드 ord(str)
- 아스키코드 -> 문자 chr(int)

```python
def solution(s, n):
    s = list(s)
    for i, x in enumerate(s):
        if x == ' ':
            continue
        else:
            if x.isupper() and ord(x) + n > ord('Z'):
                s[i] = chr(ord(x) + n - ord('Z') + ord('A') - 1)
            elif x.islower() and ord(x) + n > ord('z'):
                s[i] = chr(ord(x) + n - ord('z') + ord('a') - 1)
            else:
                s[i] = chr(ord(x) + n)
    return ''.join(s)
```

## 22. 3진법 뒤집기


### 상위 코드
- divmod() : 몫과 나머지를 같이 반환하는 함수
- X진법 문제에 약하다 개념부터 다시 쌓자

```python
def solution(n):
    answer = ''

    while n >= 1:
        rest = n % 3
        print(n,'을 3으로 나눈 나머지는',rest)
        answer += str(rest)
        n //= 3

    answer = int(answer, 3)

    return answer
```
