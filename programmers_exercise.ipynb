## 1. 크레인 인형뽑기 게임
- 단순 구현 문제

```python
def solution(board, moves):
    answer = 0
    basket = []
    
    for m in moves:
        j = m - 1
        for i in range(len(board)):
            if board[i][j] != 0:
                if len(basket) >= 1 and basket[-1] == board[i][j]:
                    answer += 2
                    basket.pop()
                else:
                    basket.append(board[i][j])
                board[i][j] = 0
                break
            
    return answer
```

## 2. 두 개 뽑아서 더하기


```python
def solution(numbers):
    answer = []
    for i in range(len(numbers) - 1):
        for j in range(i + 1, len(numbers)):
            answer.append(numbers[i] + numbers[j])
            
    answer = list(set(answer))
    answer.sort()
    return answer
```

## 3. 신규 아이디 추천
- 정규표현식과 문자열 다루기 정확히 알고 있어야
- 쉬운 것 같지만 까다로움

```python
import re

def solution(new_id):
    answer = ''
    # step1
    new_id = new_id.lower()
    # step2
    new_id = re.sub('[\~\!\@\#\$\%\^\&\*\(\)\=\+\[\{\]\}\:\?\,\<\>\/]', '', new_id)
    # step3
    new_id = re.sub('[.]{2,1000}', '.', new_id)
    # step4
    new_id = new_id.strip('.')
    # step5
    if new_id == '':
        new_id = "a"
    # step6
    if len(new_id) >= 16:
        new_id = new_id[:15]
        new_id = new_id.strip('.')
    # step7
    if len(new_id) <= 2:
        while len(new_id) <= 2:
            new_id += new_id[-1]
    answer = new_id
    return answer
```

### 상위 코드 1

```python
import re

def solution(new_id):
    st = new_id
    st = st.lower()
    st = re.sub('[^a-z0-9\-_.]', '', st)
    st = re.sub('\.+', '.', st)
    st = re.sub('^[.]|[.]$', '', st)
    st = 'a' if len(st) == 0 else st[:15]
    st = re.sub('^[.]|[.]$', '', st)
    st = st if len(st) > 2 else st + "".join([st[-1] for i in range(3-len(st))])
    return st
```

## 7. 같은 숫자는 싫어


### 내 코드
- 리스트 -> 문자열 -> 정규표현식

```python
import re

def solution(arr):
    answer = ''.join([str(i) for i in arr])
    for i in range(0, 10):
        pattern = str(i) + '+'
        answer = re.sub(pattern, str(i), answer)
    return [int(i) for i in answer]
```

## 8. 두 정수 사이의 합


### 내 풀이
- abs는 잘 썼는데 더 일반화시킬 수 있었다

```python
def solution(a, b):
    if a == b:
        return a
    # 더하는 숫자가 짝수개인 경우
    if (abs(a - b) + 1) % 2 == 0:
        return (a + b) * (abs(a - b) + 1) / 2
    # 더하는 숫자가 홀수개인 경우
    if (abs(a - b) + 1) % 2 != 0:
        return (a + b) * abs(a - b) / 2 + (a + b) / 2
```

## 10. 문자열 내 p와 y의 개수


### 내 코드
- 단순 포문

```python
def solution(s):
    s = s.lower()
    p_count, y_count = 0, 0
    for x in list(s):
        if x == 'p': p_count += 1
        if x == 'y': y_count += 1
    return True if p_count == y_count else False
```

### 상위 코드 1
- count

```python
def numPY(s):
    return s.lower().count('p') == s.lower().count('y')
```

### 상위 코드 2
- counter 쓸거면 이렇게!

```python
from collections import Counter
def numPY(s):
    c = Counter(s.lower())
    return c['y'] == c['p'] 
```

## 14. 약수의 합


### 상위 코드

```python
def solution(num):
    # num / 2 의 수들만 검사하면 성능 약 2배 향상잼
    # 자기 자신은 무조건 약수로 가짐
    # 자기 자신보다 하나 더 작은 약수는 2로 나누어 떨어지는 경우 발생
    return num + sum([i for i in range(1, (num // 2) + 1) if num % i == 0])
```

## 15. 제일 작은 수 제거하기


### 내 코드

```python
def solution(arr):
    if len(arr) == 1:
        return [-1]
    min_value = min(arr)
    for i, x in enumerate(arr):
        if x == min_value:
            arr.pop(i)
            return arr
```
