## 1. 크레인 인형뽑기 게임
- 단순 구현 문제

```python
def solution(board, moves):
    answer = 0
    basket = []
    
    for m in moves:
        j = m - 1
        for i in range(len(board)):
            if board[i][j] != 0:
                if len(basket) >= 1 and basket[-1] == board[i][j]:
                    answer += 2
                    basket.pop()
                else:
                    basket.append(board[i][j])
                board[i][j] = 0
                break
            
    return answer
```

## 2. 두 개 뽑아서 더하기


```python
def solution(numbers):
    answer = []
    for i in range(len(numbers) - 1):
        for j in range(i + 1, len(numbers)):
            answer.append(numbers[i] + numbers[j])
            
    answer = list(set(answer))
    answer.sort()
    return answer
```

## 3. 신규 아이디 추천
- 정규표현식과 문자열 다루기 정확히 알고 있어야
- 쉬운 것 같지만 까다로움

```python
import re

def solution(new_id):
    answer = ''
    # step1
    new_id = new_id.lower()
    # step2
    new_id = re.sub('[\~\!\@\#\$\%\^\&\*\(\)\=\+\[\{\]\}\:\?\,\<\>\/]', '', new_id)
    # step3
    new_id = re.sub('[.]{2,1000}', '.', new_id)
    # step4
    new_id = new_id.strip('.')
    # step5
    if new_id == '':
        new_id = "a"
    # step6
    if len(new_id) >= 16:
        new_id = new_id[:15]
        new_id = new_id.strip('.')
    # step7
    if len(new_id) <= 2:
        while len(new_id) <= 2:
            new_id += new_id[-1]
    answer = new_id
    return answer
```

### 상위 코드 1

```python
import re

def solution(new_id):
    st = new_id
    st = st.lower()
    st = re.sub('[^a-z0-9\-_.]', '', st)
    st = re.sub('\.+', '.', st)
    st = re.sub('^[.]|[.]$', '', st)
    st = 'a' if len(st) == 0 else st[:15]
    st = re.sub('^[.]|[.]$', '', st)
    st = st if len(st) > 2 else st + "".join([st[-1] for i in range(3-len(st))])
    return st
```

```python
def solution(a, b):
    month_days = [31, 29, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
    days_sum = sum(month_days[:a - 1]) + b
    weeks = ['FRI', 'SAT', 'SUN', 'MON', 'TUE', 'WED', 'THU']
    
    answer = weeks[days_sum % 7 - 1]
    
    return answer
```

## 5. 가운데 글자 가져오기


```python
def solution(s):
    if len(s) % 2 == 0:
        answer = s[len(s) // 2 - 1: len(s) // 2 + 1]
    else:
        answer = s[len(s) // 2]
    return answer
```
