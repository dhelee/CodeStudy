### 예제 4-1 상하좌우
계획서가 주어졌을 때 여행가가 최종적으로 도착할 지점의 좌표를 출력하라

```python
# 입력 조건
# 첫째 줄에 공간의 크기를 나타내는 N이 주어짐
# 둘째 줄에 여행가가 이동할 계획서 내용이 주어짐
n = int(input())
plans = input().split()
# 시작 지점
x, y = 1, 1
# 계획서 내용에 따른 이동 방향
move_types = ['L', 'R', 'U', 'D']
dx = [0, 0, -1, 1]
dy = [-1, 1, 0, 0]


# 이동 계획을 하나씩 확인함
for plan in plans:
    # 이동 후 좌표 하나씩 구하기
    for i in range(len(move_types)):
        if plan == move_types[i]:
            nx = x + dx[i]
            ny = y + dy[i]
    # 공간을 벗어나는 경우 무시
    if nx < 1 or nx > n or ny < 1 or ny > n:
        continue
    # 벗어나지 않는 경우 이동 수행
    x, y = nx, ny
    
print(x, y)
```

### 예제 4-2 시각
정수 H가 입력되면 00시 00분 00초부터 H시 59분 59초까지의 모든 시각 중에서 3이 하나라도 입력되는 모든 경우의 수 구하기  
- 모든 경우의 수를 하나씩 세서 풀 수 있는 완전 탐색 문제
- 비효율적인 시간 복잡도를 가지므로 데이터 개수가 큰 경우에는 X
- 탐색해야 할 전체 데이터 개수가 100만개 이하일 때 완전 탐색 사용이 적절

```python
# 정수 H 입력받기
h = int(input())

count = 0
# 시
for i in range(h + 1):
    # 분
    for j in range(60):
        for k in range(60):
            # 매 시각 안에 3이 포함되어 있다면 카운트 증가
            if '3' in str(i) + str(j) + str(k):
                count += 1
                
print(count)
```

### 실전문제 1 왕실의 나이트  
체스판 위 나이트의 위치가 주어졌을 때 이동할 수 있는 경우의 수를 출력하라  
- 나이트가 이동할 수 있는 2가지 방법
    1. 수평으로 두 칸 이동한 뒤에 수직으로 한 칸 이동
    2. 수직으로 두 칸 이동한 뒤에 수평으로 한 칸 이동
- 체스판 위를 벗어날 수 없음

```python
# 행 위치는 1~8까지, 열 위치는 a~h까지 표현
# 입력은 열 + 행
# 현재 나이트의 위치 입력받기
input_data = input()
# 행
row = int(input_data[1])
# 열
column = int(ord(input_data[0])) - int(ord('a')) + 1

# 나이트가 이동할 수 있는 8가지 방향 정의
steps = [(-2, -1), (-2, 1), (2, -1), (2, 1), (1, 2), (1, -2), (-1, 2), (-1, -2)]

# 8가지 방향에 대해 각 위치로 이동이 가능한지 확인
result = 0
for step in steps:
    next_row = row + step[0]
    next_column = column + step[1]
    if next_row >= 1 and next_row <= 8 and next_column >= 1 and next_column <= 8:
        result += 1
        
print(result)
```

### 실전문제 2 게임 개발
- 캐릭터는 1 X 1 크기의 정사각형으로 이루어진 N X M 크기의 직사각형 맵을 움직임
- 맵의 각 칸은 (A, B) 즉 A는 북쪽으로부터 떨어진 칸의 개수, B는 서쪽으로부터 떨어진 칸의 개수
- 캐릭터의 움직임은 다음 메뉴얼을 따름, 바다로 되어있는 공간에는 갈 수 없다
    1. 현재 위치에서 현재 방향을 기준으로 왼쪽 방향부터 차례로 갈 곳을 정함
    2. 바로 왼쪽 방향에 아직 가보지 않은 칸이 존재하면, 왼쪽 방향으로 회전한 다음 한 칸 전진. 왼쪽 방향에 가보지 않은 칸이 없다면, 왼쪽 방향으로 회전만 수행 후 1단계로 돌아감
    3. 네 방향 모두 이미 가본 칸이거나 바다로 되어있는 경우에는 바라보는 방향을 유지한 채 뒤로 한 칸 간 후에 1단계로 돌아감
    4. 뒤쪽 방향이 바다인 칸이라 돌아갈 수 없는 경우 움직임을 멈춤
- 메뉴얼을 따라 캐릭터를 이동시켰을 때 캐릭터가 방문한 칸의 수를 출력하라

```python
# 입력 조건
# 첫째 줄에 맵의 세로 크기 N과 가로 크기 M을 공백으로 구분하여 입력함 
n, m = map(int, input().split())

# 방문한 위치를 저장하기 위한 2차원 리스트(맵)를 생성하여 0으로 초기화
d = [[0] * m for _ in range(n)]

# 둘째 줄에 게임 캐릭터가 있는 칸의 좌표(A, B)와 바라보는 방향 d(북-0, 동-1, 남-2, 서-3)가 공백으로 구분하여 주어짐
# 현재 캐릭터의 위치 및 방향 입력받기
x, y, direction = map(int, input().split())
d[x][y] = 1 # 현재 좌표 방문 처리

# 셋째 줄에 맵이 육지(0)인지 바다(1)인지에 대한 정보가 주어짐, 맵의 외곽은 항상 바다로 되어있음
# 전체 맵 정보를 입력받기
map_array = []
for i in range(n):
    map_array.append(list(map(int, input().split())))
    
# 방향 이동 리스트 (북, 동, 남, 서)
dx = [-1, 0, 1, 0]
dy = [0, 1, 0, -1]

# 캐릭터가 현재 바라보는 방향에서 왼쪽으로 회전시키는 함수
# def turn_left(direction):
#     direction -= 1
#     if direction == -1:
#         direction = 3
#     return direction

def turn_left():
    global direction
    direction -= 1
    if direction == -1:
        direction = 3
        
# 시뮬레이션 시작
count = 1
# 방향 회전 횟수 초기화 - 4면 모두 갈 수 없을 경우를 확인하기 위해 필요함
turn_time = 0
while True:
    # 현재 바라보는 방향에서 왼쪽으로 회전
    # direction = turn_left(direction)
    turn_left()
    nx = x + dx[direction]
    ny = y + dy[direction]
    # 회전 후 정면에 가보지 않은 칸이 존재하는 경우 이동
    # 가보지 않은 칸이 육지이어야 함
    if d[nx][ny] == 0 and map_array[nx][ny] == 0:
        d[nx][ny] = 1
        x = nx
        y = ny
        count += 1
        # 다시 1단계 수행
        turn_time = 0
        continue
    # 정면에 가보지 않은 칸이 없거나 바다인 경우
    else:
        turn_time += 1
    # 4면 모두 갈 수 없는 경우 - 뒤로 한 칸 이동 or 게임 종료
    if turn_time == 4:
        nx = x - dx[direction]
        ny = y - dy[direction]
        # 뒤가 바다가 아닌지 확인하고 뒤로 이동
        if map_array[nx][ny] == 0:
            x = nx
            y = ny
        else:
            break
        turn_time = 0
            
print(count)    
```
