### 예제 3-1 거스름돈
- 손님에게 거슬러 줘야 할 돈이 N원일 때 거슬러 줘야 할 동전의 최소 개수?
- '가장 큰 화폐 단위부터' 돈을 거슬러 준다

```python
n = 1260
count = 0

coin_types = [500, 100, 50, 10]

for coin in coin_types:
    # 해당 화폐로 거슬러 줄 수 있는 동전의 개수 세기
    count += n // coin # 몫만 알면 된다
    
    n %= coin # 나머지만 알면 된다
    
print(count)
```

### 실전문제 1 큰 수의 법칙
- 다양한 수로 이루어진 배열을 입력 받았을 때 주어진 수들을 M번 더해 가장 큰 수를 만들어라
- 배열의 특정 인덱스에 해당하는 수가 연속해서 K번을 초과하여 더해질 수 없다
- 인덱스가 다르다면 같은 수라도 서로 다른 것으로 간주한다

1. 단순하게 풀기

```python
import time

# 입력 조건
# 첫째 줄에 N, M, K의 자연수가 주어짐 각 자연수는 공백으로 구분
n, m, k = map(int, input().split())
# 둘째 줄에 N개의 자연수가 주어짐 각 자연수는 공백으로 구분
data = list(map(int, input().split()))
# K는 항상 M보다 작거나 같음 

start_time = time.time()

# 입력 받은 수 정렬하기
data.sort(reverse=True)
# 가장 큰 수 
first = data[0]
second = data[1]

result = 0

while True:
    for i in range(k): # 가장 큰 수 k번 더하기
        if m == 0: # m이 0이라면 반복문 탈출
            break
        result += first
        m -= 1 # 더할 때마다 1씩 감소
    if m == 0: 
        break
    result += second # 두번째로 큰 수 한번 더하기
    m -= 1

    
end_time = time.time()

print(result)
print('수행 시간:', end_time - start_time)
```

2. 수열을 이용해 풀기

```python
import time

# 입력 조건
# 첫째 줄에 N, M, K의 자연수가 주어짐 각 자연수는 공백으로 구분
n, m, k = map(int, input().split())
# 둘째 줄에 N개의 자연수가 주어짐 각 자연수는 공백으로 구분
data = list(map(int, input().split()))
# K는 항상 M보다 작거나 같음 

start_time = time.time()

# 입력 받은 수 정렬하기
data.sort(reverse=True)
# 가장 큰 수 
first = data[0]
second = data[1]

# 가장 큰 수가 더해지는 횟수 계산하기
count = int(m / (k+1)) * k + m % (k+1)

result = 0
result += count * first
result += (m-count) * second

end_time = time.time()

print(result)
print('수행 시간:', end_time - start_time)
```
