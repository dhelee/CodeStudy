## 기출문제 1 모험가 길드
공포도가 X인 모험가는 반드시 X명 이상으로 구성한 모험가 그룹에 참여해야 여행을 떠날 수 있다. 최대 몇 개의 모험가 그룹을 만들 수 있는가?

### 내 풀이

```python
# 입력 조건
# 첫째 줄에 모험가의 수 N이 주어짐
n = int(input())
# 둘쨰 줄에 각 모험가의 공포도 값이 주어짐 
x = list(map(int, input().split()))

# 공포도 리스트 오름차순 정렬
x.sort()

# 정답 저장할 변수
result = 0
# 길드원 수를 세기 위한 변수 
count = 1

for i in range(n):
    # 공포도가 현재 길드원 수보다 작거나 같아야 함 
    if x[i] <= count:
        result += 1
        count = 1
    else:
        count += 1
        
print(result)
```

### 정답 코드

```python
# 입력 조건
# 첫째 줄에 모험가의 수 N이 주어짐
n = int(input())
# 둘쨰 줄에 각 모험가의 공포도 값이 주어짐 
x = list(map(int, input().split()))

# 공포도 리스트 오름차순 정렬
x.sort()

# 정답 저장할 변수
result = 0
# 길드원 수를 세기 위한 변수 
count = 0

for i in x:
    count += 1 # 현재 그룹에 해당 모험가를 포함시킴
    if count >= i: # 현재 그룹에 포함된 모험가 수가 현재의 공포도 이상일 때 그룹 결성
        result += 1
        count = 0

print(result)
```

## 기출문제 2 곱하기 혹은 더하기
각 자리가 숫자로만 이루어진 문자열이 주어졌을 때, 숫자 사이에 X 혹은 + 연산자를 넣어 결과적으로 만들어질 수 있는 가장 큰 수를 구하라.

### 내 풀이

```python
# 입력 조건
# 첫째 줄에 여러 개의 숫자로 구성된 하나의 문자열 S가 주어짐
s = input()

# 연산을 시작할 숫자
result = int(s[0])

for i in range(1, len(s)):
    if int(s[i]) <= 1 or result <= 1:
        result += int(s[i])
    else:
        result *= int(s[i])
        
print(result)
```

### 정답 코드

```python
# 입력 조건
# 첫째 줄에 여러 개의 숫자로 구성된 하나의 문자열 S가 주어짐
s = input()

# 연산을 시작할 숫자
result = int(s[0])

for i in range(1, len(s)):
    num = int(s[i])
    if num <= 1 or result <= 1:
        result += num
    else:
        result *= num
        
print(result)
```

## 기출문제 3 문자열 뒤집기
0과 1로만 이루어진 문자열 S가 주어졌을 때, 연속된 하나 이상의 숫자를 뒤집어서 문자열의 모든 숫자를 전부 같게 만들어라. 최소 뒤집는 횟수는?

### 정답 코드

```python
# 입력 조건
# 첫째 줄에 0과 1로만 이루어진 문자열 S가 주어짐
s = input()

# 모두 0으로 만드는 경우와 모두 1로 만드는 경우를 비교
count0 = 0
count1 = 0

# 첫번째 원소에 대해 처리
if s[0] == '1':
    count0 += 1
else:
    count1 += 1
    
# 두번째 원소부터 모든 원소 확인
for i in range(len(s) - 1):
    if s[i] != s[i + 1]:
        # 다음 수에서 1로 바뀌는 경우
        if s[i + 1] == '1':
            count0 += 1
        # 다음 수에서 0으로 바뀌는 경우 
        else:
            count1 += 1
            
print(min(count0, count1))
```

## 기출문제 4 만들 수 없는 금액
N개의 동전을 이용하여 만들 수 없는 양의 정수 금액 중 최솟값을 구하는 프로그램 작성

### 정답 코드

```python
# 입력조건
# 첫째 줄에는 동전의 개수를 나타내는 N이 주어짐
n = int(input())
# 둘째 줄에는각 동전의 화폐 단위를 나타내는 N개 자연수가 주어짐
coins = list(map(int, input().split()))

# 동전의 화폐 단위를 오름차순 정렬
coins.sort()

target = 1
for coin in coins:
    if target < coin:
        break
    target += coin

# 만들 수 없는 금액 출력
print(target)
```

## 기출문제 5 볼링공 고르기
두 사람이 서로 다른 무게의 볼링공을 고르려고 할 때, 고를 수 있는 경우의 수를 구하라

### 내 풀이

```python
# 입력 조건
# 첫째 줄에 볼링공 개수 N과 공의 최대 무게 M이 주어짐
n, m = map(int, input().split())
# 둘째 줄에 각 볼링공의 무게가 공백으로 구분되어 주어짐
weights = list(map(int, input().split()))

# 무게 오름차순 정렬
weights.sort()

result = 0

for i in range(0, n - 1):
    # 첫번째 공을 고르고 시작
    selected = weights[i]
    for j in range(i + 1, n):
        if selected != weights[j]:
            result += 1
            
print(result)
```

### 정답 코드
단계가 진행됨에 따라 B가 선택하는 경우의 수가 줄어든다! (조합)  
각 무게별로 볼링공이 몇 개씩 존재하는지 기록하는 리스트 변수 선언

```python
n, m = map(int, input().split())
data = list(map(int, input().split()))

# 1부터 10까지의 무게를 담을 수 있는 리스트
array = [0] * 11

for x in data:
    array[x] += 1
    
result = 0
# 1부터 m까지의 각 무게에 대하여 처리
for i in range(1, m + 1):
    n -= array[i] # 무게가 i인 볼링공의 개수(A가 선택할 수 있는 개수) 제외
    result += array[i] * n # B가 선책하는 경우의 수와 곱하기
    
print(result)
```

## 기출문제 6 무지의 먹방 라이브
무지는 N개 음식을 1번부터 순서대로 1초씩 섭취한다.  
각 음식을 모두 먹는 데 필요한 시간이 담긴 배열과, 네트워크 장애가 발생한 시간 K초가 매개변수로 주어질 때 몇 번 음식부터 다시 섭취하면 되는지 구하라.

### 내 풀이 - 잘못된 접근 ㅜ ㅜ

```python
def solution(food_times, k):
    answer = 0
    n = len(food_times)
    flag = True
    
    # k가 전체 음식 수보다 작거나 같을 경우
    if k <= n:
        answer = k + 1
    # k가 전체 음식 수보다 클 경우 (모든 음식을 한번 이상 섭취)
    else:
        while flag:
            # 한바퀴 돌고 난 후 남은 시간
            k -= n
            if max(food_times) == 1:
                answer = -1
                break
            for i in range(n):
                food_times[i] = food_times[i] - 1
                if food_times[i] != 0:
                    k -= 1
                    food_times[i] = food_times[i] - 1
                if k == 0:
                    answer = i + 1
                    if answer == n: answer = 1
                    flag = False
                
    return answer
```

```python
print(solution([3, 1, 2], 5))
```

### 정답 코드
시간이 적게 걸리는 음식부터 확인!  
모든 음식을 시간 기준으로 정렬한 뒤, 시간이 적게 걸리는 음식부터 제거  
우선순위 큐를 이용하여 구현

```python
import heapq

def solution(food_times, k):
    # 전체 음식을 먹는 시간보다 k가 크거나 같다면 -1
    if sum(food_times) <= k:
        return -1
    
    # 시간이 작은 음식부터 빼야 하므로 우선순위 큐 이용
    q = []
    for i in range(len(food_times)):
        # (음식 시간, 음식 번호) 형태로 우선순위 큐에 삽입
        heapq.heappush(q, (food_times[i], i + 1))
        
    sum_value = 0 # 먹기 위해 사용한 시간
    previous = 0 # 직전에 다 먹은 음식 시간
    
    length = len(food_times) # 남은 음식 개수
    
    # sum_value + (현재의 음식 시간 - 이전 음식 시간) * 현재 음식 개수와 k 비교
    while sum_value + ((q[0][0] - previous) * length) <= k:
        now = heapq.heappop(q)[0]
        sum_value += (now - previous) * length
        length -= 1 # 다 먹은 음식 제외
        previous = now # 이전 음식 시간 재설정
        
    # 남은 음식 중에서 몇 번째 음식인지 확인하여 출력
    result = sorted(q, key=lambda x: x[1]) # 음식의 번호 기준으로 정렬
    return result[(k - sum_value) % length][1]
    
```
